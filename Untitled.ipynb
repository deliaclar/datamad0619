```python
def cutToNum(x):
    if 'Ideal'in str(x): return 10
    elif 'Premium' in str(x): return 8
    elif 'Very' in str(x): return 6
    elif 'Good' in str(x): return 4
    elif 'Fair' in str(x): return 2
df_tr['cutNumeric']=df_tr['cut'].apply(cutToNum)
df_tr['cutNumeric'].value_counts()
```

```python
def colorToNum(x):
    if 'D' in str(x): return 7
    if 'E' in str(x): return 6
    if 'F' in str(x): return 5
    if 'G' in str(x): return 4
    if 'H' in str(x): return 3
    if 'I' in str(x): return 2
    if 'J' in str(x): return 1

df_tr['color']=df_tr['color'].apply(colorToNum)
df_tr['color'].value_counts()
```

```python
def clarityToNum(x):
    if 'IF' in str(x): return 7
    if 'VVS1' in str(x): return 6
    if 'VVS2' in str(x): return 5
    if 'VS1' in str(x): return 4
    if 'VS2' in str(x): return 3
    if 'SI1' in str(x): return 2
    if 'SI2' in str(x): return 1
    if 'I1' in str(x): return 0

df_tr['clarity']=df_tr['clarity'].apply(clarityToNum)
df_tr['clarity'].value_counts()
```
